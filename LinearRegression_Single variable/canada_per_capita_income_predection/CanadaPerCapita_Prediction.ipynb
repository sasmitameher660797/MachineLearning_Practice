<h3>Problem Statement :</h3> <p >Predict canada's per capita income in year 2020.Use canada_per_capita_income.csv file. Using this build a regression model and predict the per capita income fo canadian citizens in year 2020</p>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
df
```

```python
df.shape
```

```python
df.columns
```

```python
df.index
```

```python
plt.scatter(df.year,df['per capita income (US$)'],c='red')
plt.xlabel('year')
plt.ylabel('per capita income(US$)')
```

```python
new_df=df.drop('per capita income (US$)',axis=1)
new_df
```

```python
income= df['per capita income (US$)']
income
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,income)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
